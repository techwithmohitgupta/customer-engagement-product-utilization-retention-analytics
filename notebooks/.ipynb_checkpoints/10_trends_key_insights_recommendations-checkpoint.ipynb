## Overall Retention and Churn Trend Summary

```python
# Step 10G: Overall retention and churn trend summary

total_customers = len(df)
churned_customers = int(df["Exited"].sum())
retained_customers = int((df["Exited"] == 0).sum())

overall_churn_rate = round(churned_customers / total_customers * 100, 2)
overall_retention_rate = round(retained_customers / total_customers * 100, 2)

avg_retention_strength_score = round(df["Retention_Strength_Score"].mean(), 2)
median_retention_strength_score = round(df["Retention_Strength_Score"].median(), 2)

avg_relationship_strength = round(df["Relationship_Strength_Index"].mean(), 2)

overall_retention_trend_summary = pd.DataFrame({
    "Metric": [
        "Total Customers",
        "Retained Customers",
        "Churned Customers",
        "Overall Retention Rate",
        "Overall Churn Rate",
        "Average Retention Strength Score",
        "Median Retention Strength Score",
        "Average Relationship Strength Index"
    ],
    "Value": [
        f"{total_customers:,}",
        f"{retained_customers:,}",
        f"{churned_customers:,}",
        f"{overall_retention_rate:.2f}%",
        f"{overall_churn_rate:.2f}%",
        f"{avg_retention_strength_score:.2f}",
        f"{median_retention_strength_score:.2f}",
        f"{avg_relationship_strength:.2f}"
    ],
    "Business Meaning": [
        "Total customer base available for retention analysis",
        "Customers who stayed with the bank",
        "Customers who exited/churned",
        "Baseline retention performance",
        "Baseline churn exposure",
        "Average customer-level retention strength",
        "Middle-point retention strength score",
        "Average relationship depth and loyalty signal"
    ]
})

overall_retention_trend_summary
```

## Retention Strength Tier Key Trend Analysis

```python
# Step 10H: Retention strength tier key trend analysis

retention_tier_trend = (
    df.groupby("Retention_Strength_Tier")
    .agg(
        Customer_Count=("CustomerId", "count"),
        Churned_Customers=("Exited", "sum"),
        Retained_Customers=("Exited", lambda x: (x == 0).sum()),
        Avg_Retention_Strength_Score=("Retention_Strength_Score", "mean"),
        Avg_Relationship_Strength=("Relationship_Strength_Index", "mean"),
        Avg_Num_Products=("NumOfProducts", "mean"),
        Avg_Balance=("Balance", "mean")
    )
    .reset_index()
)

retention_tier_trend["Customer_%"] = (
    retention_tier_trend["Customer_Count"] / len(df) * 100
).round(2)

retention_tier_trend["Churn_Rate"] = (
    retention_tier_trend["Churned_Customers"] /
    retention_tier_trend["Customer_Count"] * 100
).round(2)

retention_tier_trend["Retention_Rate"] = (
    retention_tier_trend["Retained_Customers"] /
    retention_tier_trend["Customer_Count"] * 100
).round(2)

for col in [
    "Avg_Retention_Strength_Score",
    "Avg_Relationship_Strength",
    "Avg_Num_Products",
    "Avg_Balance"
]:
    retention_tier_trend[col] = retention_tier_trend[col].round(2)

retention_tier_order = [
    "Critical Retention Risk",
    "Weak Retention",
    "Moderate Retention",
    "Strong Retention",
    "Very Strong Retention"
]

retention_tier_trend["Retention_Strength_Tier"] = pd.Categorical(
    retention_tier_trend["Retention_Strength_Tier"],
    categories=retention_tier_order,
    ordered=True
)

retention_tier_trend = retention_tier_trend.sort_values(
    "Retention_Strength_Tier"
).reset_index(drop=True)

highest_churn_tier = retention_tier_trend.sort_values(
    "Churn_Rate",
    ascending=False
).iloc[0]

best_retention_tier = retention_tier_trend.sort_values(
    "Retention_Rate",
    ascending=False
).iloc[0]

retention_tier_trend
```

## Engagement and Product Utilization Key Trends

```python
# Step 10I: Engagement and product utilization key trends

# -----------------------------
# 1. Engagement Status Trend
# -----------------------------

engagement_trend = (
    df.groupby("Engagement_Status")
    .agg(
        Customer_Count=("CustomerId", "count"),
        Churned_Customers=("Exited", "sum"),
        Retained_Customers=("Exited", lambda x: (x == 0).sum()),
        Avg_Retention_Strength_Score=("Retention_Strength_Score", "mean"),
        Avg_Relationship_Strength=("Relationship_Strength_Index", "mean"),
        Avg_Num_Products=("NumOfProducts", "mean")
    )
    .reset_index()
)

engagement_trend["Churn_Rate"] = (
    engagement_trend["Churned_Customers"] /
    engagement_trend["Customer_Count"] * 100
).round(2)

engagement_trend["Retention_Rate"] = (
    engagement_trend["Retained_Customers"] /
    engagement_trend["Customer_Count"] * 100
).round(2)

for col in [
    "Avg_Retention_Strength_Score",
    "Avg_Relationship_Strength",
    "Avg_Num_Products"
]:
    engagement_trend[col] = engagement_trend[col].round(2)

engagement_trend_clean = engagement_trend.rename(
    columns={"Engagement_Status": "Segment_Name"}
)

engagement_trend_clean["Trend_Area"] = "Engagement Status"


# -----------------------------
# 2. Product Depth Segment Trend
# -----------------------------

product_depth_trend = (
    df.groupby("Product_Depth_Segment")
    .agg(
        Customer_Count=("CustomerId", "count"),
        Churned_Customers=("Exited", "sum"),
        Retained_Customers=("Exited", lambda x: (x == 0).sum()),
        Avg_Retention_Strength_Score=("Retention_Strength_Score", "mean"),
        Avg_Relationship_Strength=("Relationship_Strength_Index", "mean"),
        Avg_Num_Products=("NumOfProducts", "mean")
    )
    .reset_index()
)

product_depth_trend["Churn_Rate"] = (
    product_depth_trend["Churned_Customers"] /
    product_depth_trend["Customer_Count"] * 100
).round(2)

product_depth_trend["Retention_Rate"] = (
    product_depth_trend["Retained_Customers"] /
    product_depth_trend["Customer_Count"] * 100
).round(2)

for col in [
    "Avg_Retention_Strength_Score",
    "Avg_Relationship_Strength",
    "Avg_Num_Products"
]:
    product_depth_trend[col] = product_depth_trend[col].round(2)

product_depth_trend_clean = product_depth_trend.rename(
    columns={"Product_Depth_Segment": "Segment_Name"}
)

product_depth_trend_clean["Trend_Area"] = "Product Depth Segment"


# -----------------------------
# 3. Combine Both Trend Tables Safely
# -----------------------------

engagement_product_key_trends = pd.concat(
    [
        engagement_trend_clean,
        product_depth_trend_clean
    ],
    ignore_index=True
)

engagement_product_key_trends = engagement_product_key_trends[
    [
        "Trend_Area",
        "Segment_Name",
        "Customer_Count",
        "Churned_Customers",
        "Retained_Customers",
        "Churn_Rate",
        "Retention_Rate",
        "Avg_Retention_Strength_Score",
        "Avg_Relationship_Strength",
        "Avg_Num_Products"
    ]
]

engagement_product_key_trends
```

## Customer Risk and Relationship Strength Key Trends

```python
# Step 10J: Customer risk and relationship strength key trends

customer_risk_trend = (
    df.groupby("Customer_Risk_Segment")
    .agg(
        Customer_Count=("CustomerId", "count"),
        Churned_Customers=("Exited", "sum"),
        Retained_Customers=("Exited", lambda x: (x == 0).sum()),
        Avg_Retention_Strength_Score=("Retention_Strength_Score", "mean"),
        Avg_Relationship_Strength=("Relationship_Strength_Index", "mean"),
        Avg_Balance=("Balance", "mean"),
        Avg_Num_Products=("NumOfProducts", "mean")
    )
    .reset_index()
)

customer_risk_trend["Customer_%"] = (
    customer_risk_trend["Customer_Count"] / len(df) * 100
).round(2)

customer_risk_trend["Churn_Rate"] = (
    customer_risk_trend["Churned_Customers"] /
    customer_risk_trend["Customer_Count"] * 100
).round(2)

customer_risk_trend["Retention_Rate"] = (
    customer_risk_trend["Retained_Customers"] /
    customer_risk_trend["Customer_Count"] * 100
).round(2)

for col in [
    "Avg_Retention_Strength_Score",
    "Avg_Relationship_Strength",
    "Avg_Balance",
    "Avg_Num_Products"
]:
    customer_risk_trend[col] = customer_risk_trend[col].round(2)

customer_risk_trend = customer_risk_trend.sort_values(
    "Churn_Rate",
    ascending=False
).reset_index(drop=True)


relationship_strength_trend = (
    df.groupby("Relationship_Strength_Tier")
    .agg(
        Customer_Count=("CustomerId", "count"),
        Churned_Customers=("Exited", "sum"),
        Retained_Customers=("Exited", lambda x: (x == 0).sum()),
        Avg_Retention_Strength_Score=("Retention_Strength_Score", "mean"),
        Avg_Relationship_Strength=("Relationship_Strength_Index", "mean"),
        Avg_Balance=("Balance", "mean"),
        Avg_Num_Products=("NumOfProducts", "mean")
    )
    .reset_index()
)

relationship_strength_trend["Churn_Rate"] = (
    relationship_strength_trend["Churned_Customers"] /
    relationship_strength_trend["Customer_Count"] * 100
).round(2)

relationship_strength_trend["Retention_Rate"] = (
    relationship_strength_trend["Retained_Customers"] /
    relationship_strength_trend["Customer_Count"] * 100
).round(2)

for col in [
    "Avg_Retention_Strength_Score",
    "Avg_Relationship_Strength",
    "Avg_Balance",
    "Avg_Num_Products"
]:
    relationship_strength_trend[col] = relationship_strength_trend[col].round(2)

highest_risk_segment = customer_risk_trend.iloc[0]
highest_relationship_churn = relationship_strength_trend.sort_values(
    "Churn_Rate",
    ascending=False
).iloc[0]

customer_risk_trend
```

## Step 10K: Financial Commitment and Credit Card Key Trends

This step analyzes whether salary-balance mismatch and credit card ownership show strong retention or churn patterns.

The goal is to verify whether financial commitment indicators are useful standalone retention drivers or only supporting signals.

```python
# Step 10K: Financial commitment and credit card key trends

salary_balance_trend = (
    df.groupby("Salary_Balance_Mismatch_Status")
    .agg(
        Customer_Count=("CustomerId", "count"),
        Churned_Customers=("Exited", "sum"),
        Retained_Customers=("Exited", lambda x: (x == 0).sum()),
        Avg_Balance=("Balance", "mean"),
        Avg_Estimated_Salary=("EstimatedSalary", "mean"),
        Avg_Retention_Strength_Score=("Retention_Strength_Score", "mean"),
        Avg_Relationship_Strength=("Relationship_Strength_Index", "mean"),
        Avg_Num_Products=("NumOfProducts", "mean")
    )
    .reset_index()
)

salary_balance_trend["Churn_Rate"] = (
    salary_balance_trend["Churned_Customers"] /
    salary_balance_trend["Customer_Count"] * 100
).round(2)

salary_balance_trend["Retention_Rate"] = (
    salary_balance_trend["Retained_Customers"] /
    salary_balance_trend["Customer_Count"] * 100
).round(2)

for col in [
    "Avg_Balance",
    "Avg_Estimated_Salary",
    "Avg_Retention_Strength_Score",
    "Avg_Relationship_Strength",
    "Avg_Num_Products"
]:
    salary_balance_trend[col] = salary_balance_trend[col].round(2)


credit_card_trend = (
    df.groupby("Credit_Card_Status")
    .agg(
        Customer_Count=("CustomerId", "count"),
        Churned_Customers=("Exited", "sum"),
        Retained_Customers=("Exited", lambda x: (x == 0).sum()),
        Avg_Retention_Strength_Score=("Retention_Strength_Score", "mean"),
        Avg_Relationship_Strength=("Relationship_Strength_Index", "mean"),
        Avg_Num_Products=("NumOfProducts", "mean"),
        Avg_Balance=("Balance", "mean")
    )
    .reset_index()
)

credit_card_trend["Churn_Rate"] = (
    credit_card_trend["Churned_Customers"] /
    credit_card_trend["Customer_Count"] * 100
).round(2)

credit_card_trend["Retention_Rate"] = (
    credit_card_trend["Retained_Customers"] /
    credit_card_trend["Customer_Count"] * 100
).round(2)

for col in [
    "Avg_Retention_Strength_Score",
    "Avg_Relationship_Strength",
    "Avg_Num_Products",
    "Avg_Balance"
]:
    credit_card_trend[col] = credit_card_trend[col].round(2)


financial_commitment_key_trends = pd.concat(
    [
        salary_balance_trend.rename(
            columns={"Salary_Balance_Mismatch_Status": "Segment_Name"}
        ).assign(Trend_Area="Salary-Balance Mismatch"),
        
        credit_card_trend.rename(
            columns={"Credit_Card_Status": "Segment_Name"}
        ).assign(Trend_Area="Credit Card Ownership")
    ],
    ignore_index=True
)

financial_commitment_key_trends = financial_commitment_key_trends[
    [
        "Trend_Area",
        "Segment_Name",
        "Customer_Count",
        "Churned_Customers",
        "Retained_Customers",
        "Churn_Rate",
        "Retention_Rate",
        "Avg_Retention_Strength_Score",
        "Avg_Relationship_Strength",
        "Avg_Num_Products",
        "Avg_Balance"
    ]
]

financial_commitment_key_trends
```

## Step 10N: Customer Segment Action Plan
