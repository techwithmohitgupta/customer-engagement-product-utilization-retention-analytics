# Customer Engagement & Product Utilization Analytics for Retention Strategy

## Notebook 01: Project Overview & Data Loading

### Project Overview

This project analyzes customer engagement, product utilization, and relationship strength to understand customer retention and churn behavior in a banking context.

Banks often assume that customers with high balances, good salaries, or strong financial profiles are loyal. However, customers may still churn if they are inactive, use limited banking products, or have weak relationship depth with the bank.

The main objective of this project is to evaluate how customer activity, product usage, credit card ownership, balance, salary, tenure, and relationship strength are associated with customer retention.

---

### Business Problem

The bank wants to understand why some customers leave even when they may appear financially valuable.

The key business questions are:

1. Are active customers less likely to churn than inactive customers?
2. Does product usage depth improve customer retention?
3. Are high-balance customers always loyal?
4. Which high-value customers are disengaged and at risk?
5. Can customer relationship strength be measured using engagement and product behavior?

---

### Target Variable

The target column in this dataset is `Exited`.

- `Exited = 0`: Customer retained
- `Exited = 1`: Customer churned

## Missing Values Check

```python
# Check missing values in each column
missing_values = df.isnull().sum().reset_index()
missing_values.columns = ["Column", "Missing Values"]

missing_values["Missing %"] = (missing_values["Missing Values"] / len(df) * 100).round(2)

missing_values
```

### Total Missing Values Summary

```python
total_missing = df.isnull().sum().sum()

print(f"Total Missing Values in Dataset: {total_missing}")
```

### Duplicate Records Check

```python
duplicate_rows = df.duplicated().sum()

print(f"Total Duplicate Rows: {duplicate_rows}")
```

### Customer ID Uniqueness Check

```python
total_customers = df["CustomerId"].count()
unique_customers = df["CustomerId"].nunique()
duplicate_customer_ids = total_customers - unique_customers

print(f"Total Customer Records: {total_customers:,}")
print(f"Unique Customer IDs: {unique_customers:,}")
print(f"Duplicate Customer IDs: {duplicate_customer_ids:,}")
```

### Target Column Validation

```python
# Validate churn target column
target_distribution = df["Exited"].value_counts().reset_index()
target_distribution.columns = ["Exited", "Customer Count"]

target_distribution["Percentage"] = (
    target_distribution["Customer Count"] / len(df) * 100
).round(2)

target_distribution
```

### Churn Rate Summary

```python
retained_customers = (df["Exited"] == 0).sum()
churned_customers = (df["Exited"] == 1).sum()
overall_churn_rate = df["Exited"].mean() * 100
overall_retention_rate = 100 - overall_churn_rate

print(f"Retained Customers: {retained_customers:,}")
print(f"Churned Customers: {churned_customers:,}")
print(f"Overall Retention Rate: {overall_retention_rate:.2f}%")
print(f"Overall Churn Rate: {overall_churn_rate:.2f}%")
```

### Binary Columns Validation
